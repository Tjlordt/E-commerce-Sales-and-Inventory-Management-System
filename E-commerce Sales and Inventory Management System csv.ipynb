```python
import numpy as np
import pandas as pd

# Number of samples
n = 1000

# Nominal data: Product categories, colors
product_categories = np.random.choice(['Electronics', 'Clothing', 'Home Appliances'], n)
product_colors = np.random.choice(['Red', 'Blue', 'Green'], n)


# Ordinal data: Product ratings
product_ratings = np.random.choice(['Poor', 'Fair', 'Good', 'Excellent'], n, p=[0.1, 0.3, 0.4, 0.2])

# Interval data: Product weight, price, and quantity in stock
product_weights = np.random.uniform(50, 1000, n)
product_prices = np.random.uniform(10, 1000, n)
quantity_in_stock = np.random.randint(0, 100, n)

# Ratio data: Product IDs, Order IDs, Customer IDs, and Employee IDs
product_ids = range(1, n + 1)

# Create Products DataFrame
products = pd.DataFrame({
    'ProductID': product_ids,
    'ProductName': [f'Product {i}' for i in product_ids],
    'ProductRating': product_ratings,
    'ProductWeight': product_weights,
    'ProductPrice': product_prices,
    'ProductCategory': product_categories,
    'ProductColor': product_colors,
})

# Save Products table to a CSV file
products.to_csv('products.csv', index=False)

# Display the Products DataFrame
print("Products Table:")
print(products.head())

# Create Sales Orders DataFrame
sales_orders = pd.DataFrame({
    'SalesOrderID': range(1, n + 1),
    'OrderDate': pd.date_range(start='2023-01-01', periods=n, freq='D'),
    'CustomerID': np.random.randint(1, 11, n),
    'EmployeeID': np.random.randint(1, 11, n)
})

# Save Sales Orders table to a CSV file
sales_orders.to_csv('sales_orders.csv', index=False)

# Display the Sales Orders DataFrame
print("\nSales Orders Table:")
print(sales_orders.head())

# Create Order Details DataFrame
order_details = pd.DataFrame({
    'OrderID': np.random.choice(range(1, n + 1), n),
    'ProductID': np.random.choice(range(1, n + 1), n),
    'Quantity': np.random.randint(1, 10, n)
})

# Save Order Details table to a CSV file
order_details.to_csv('order_details.csv', index=False)

# Display the Order Details DataFrame
print("\nOrder Details Table:")
print(order_details.head())

# Create Customers DataFrame
customers = pd.DataFrame({
    'CustomerID': range(1, 11),
    'CustomerName': [f'Customer {i}' for i in range(1, 11)],
    'Email': [f'customer{i}@example.com' for i in range(1, 11)],
    'PhoneNumber': ['555-123-4567', '555-234-5678', '555-345-6789', '555-456-7890', '555-567-8901',
                    '555-678-9012', '555-789-0123', '555-890-1234', '555-901-2345', '555-012-3456']
})

# Save Customers table to a CSV file
customers.to_csv('customers.csv', index=False)

# Display the Customers DataFrame
print("\nCustomers Table:")
print(customers.head())

# Create Inventory DataFrame
inventory = pd.DataFrame({
    'ProductID': range(1, n + 1),
    'QuantityInStock': np.random.randint(0, 100, n)
})

# Save Inventory table to a CSV file
inventory.to_csv('inventory.csv', index=False)

# Display the Inventory DataFrame
print("\nInventory Table:")
print(inventory.head())

# Create Employees DataFrame
employees = pd.DataFrame({
    'EmployeeID': range(1, 11),
    'EmployeeName': [f'Employee {i}' for i in range(1, 11)],
    'Role': np.random.choice(['Manager', 'Salesperson', 'Technician'], 10)
})

# Save Employees table to a CSV file
employees.to_csv('employees.csv', index=False)

# Display the Employees DataFrame
print("\nEmployees Table:")
print(employees.head())
```
